```python
fruit_info_filtered = fruit_info[['Weight', 'LR_Diff', 'MajorDiameter', 'Volume', 'SizerSize', 
                        'RodPulse', 'RodIndex', 'SizeName', 'Red', 'Red/Orange',
                        'Dark Orange', 'Orange', 'Orange/Yellow', 'Yellow/Green', 'Green', 
                        'Scar', 'Texture', 'Blem', 'Dark Blem', 'Overall Roundness', 
                        'Trapeziod Angle (Deg)', 'Stem angle', 'Smoothness', 'Flatness', 
                        'Elongation', 'Calyx size', 'Stem Size', 'IQS 1 avg', 'IQS 2 avg', 
                        'Fruit', 'Rough skin', 'Fruit Center X (mm)', 'Fruit Center Y (mm)']]
correlation_matrix = fruit_info_filtered.astype('float64').corr()
correlation_matrix
```

* The heuristic is to drop aa group of high-correlated columns and to keep the last one.
* This might not be effective for feature selection.
* A comprehensive factor analysis should be preformed.

```python
for col in correlation_matrix.columns :
    if correlation_matrix[col].between(0.8, 1.0, inclusive = False).any() or \
       correlation_matrix[col].between(-1.0, -0.8).any():
        correlation_matrix = correlation_matrix.drop(col)
        correlation_matrix = correlation_matrix.drop(col, axis = 1)
print("shape : " + str(correlation_matrix.shape))
correlation_matrix
```

```python
fruit_info_filtered = fruit_info[correlation_matrix.columns]
fruit_info_filtered = fruit_info_filtered.rename(index = str, columns = {"IQS 1 avg" : "Brix", "IQS 2 avg" : "granulation"})
print('Shape : ' + str(fruit_info_filtered.shape))
fruit_info_filtered.head()
```
